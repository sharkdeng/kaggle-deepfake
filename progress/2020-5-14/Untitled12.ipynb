```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, Dense, Dropout, MaxPooling2D
from tensorflow.keras.layers import Input, GlobalAveragePooling2D
from tensorflow.keras.models import Sequential, Model
from  tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import CategoricalCrossentropy

from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np

y_true = [[0, 1, 0], [0, 0, 1]]
y_pred = [[0.05, 0.95, 0], [0.1, 0.8, 0.1]]
# Using 'auto'/'sum_over_batch_size' reduction type.
cce = tf.keras.losses.CategoricalCrossentropy()
cce(y_true, y_pred).numpy()
```

```python
(train_x, train_y), (test_x, test_y) = mnist.load_data()
train_x = tf.convert_to_tensor(train_x, dtype = tf.float32) / 255.
train_y =tf.convert_to_tensor(train_y, dtype = tf.int32)
train_y  = tf.one_hot(train_y, depth = 10)
train_ds = tf.data.Dataset.from_tensor_slices((train_x, train_y)).batch(32).shuffle(10000)
for step, (x, y) in enumerate(train_ds):
    print(x.shape, y.shape)
    img = x[0] * 255
    img = img.numpy().astype(np.int)
    plt.imshow(img)
    break
```

```python
def build_model():
    model = Sequential(
    [
        Input(shape=(28, 28, 1)),
        Conv2D(32, kernel_size=2, strides=2),
        Conv2D(16, kernel_size=2, strides=2),
        GlobalAveragePooling2D(),
        Dense(64, activation='relu'),
        Dense(10, activation='softmax')
    ])
    
    return model
```

```python
model = build_model()
model.summary()
```

```python
loss_fn = CategoricalCrossentropy()
opt = Adam(lr=0.0001)
bias = []
losses = []
steps = []
for step, (x, y) in enumerate(train_ds):
    x = tf.expand_dims(x, 3)
  
    with tf.GradientTape() as tape:
        tape.watch(x)
        pred = model(x)
        loss = loss_fn(y, pred)
        print("loss=", loss.numpy())
        losses.append(loss.numpy())
        
        
        
    grads = tape.gradient(loss, model.trainable_variables)
    opt.apply_gradients(zip(grads,model.trainable_variables))
    var = model.trainable_variables

    bias.append(var[0][1])

```
